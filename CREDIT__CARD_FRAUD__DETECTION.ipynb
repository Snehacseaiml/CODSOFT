importing the dependencies


```python
# Importing commonly used libraries with custom aliases
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to the padas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the no of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fradulent transactions
credit_card_data['Class'].value_counts()
```

0 is normal transaction
1 is fradulent transaction

```python
#seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#some statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

building a sample dataset containing a similar distribution of normal transaction and fraudulent transaction


no of fraudulent transaction is 492


```python
legit_sample = legit.sample(n=492)
```

concatenating two dataframe

```python
new_dataset= pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting the data into features and targets

```python
X= new_dataset.drop(columns='Class', axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

split the data into traning data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model training

logistic regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10000)  # or try 2000
model.fit(X_train, Y_train)
```

MODEL EVALUATION

ACCURACY SCORE

```python
#ACCURACY ON THE TRAINING DATA
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('accuracy on traing data:', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction =  model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('accuracy on test data:', test_data_accuracy)
```
